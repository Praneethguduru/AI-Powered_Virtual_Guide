```python
from bs4 import BeautifulSoup
import requests
```

```python
bs = BeautifulSoup()
response = requests.get('https://www.hyderabadtourism.in/hyderabad-sightseeing')
bs = BeautifulSoup(response.text, 'html.parser')
print(bs.prettify())
```

```python
paragraph_tags = bs.find_all('p')
print(paragraph_tags)

```

```python
paragraph_texts = [tag.text for tag in paragraph_tags]
print(paragraph_texts)
```

```python
for paragraph in paragraph_texts:
    print(paragraph)

```
