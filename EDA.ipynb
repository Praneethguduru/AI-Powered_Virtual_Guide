```python
places_count = places['City'].value_counts()
print(places_count)
```

```python
#Ploting the number of places in each city
places_count.plot(kind='bar', figsize=(20, 6), color='red')
plt.title('Number of places in each city')
plt.xlabel('City')
plt.ylabel('Number of places')
```

```python
#Plotiing rating of places in each city
places.groupby('City')['Ratings'].mean().plot(kind='bar', figsize=(20, 6), color='blue')
plt.title('Average rating of places in each city')
plt.xlabel('City')
plt.ylabel('Average rating')
plt.show()
```
